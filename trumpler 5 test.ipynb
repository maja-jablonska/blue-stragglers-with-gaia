```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
gaiadr3 = pd.read_csv('./Trumpler 5.csv')
lit = pd.read_csv('./Trumpler 5_literature.csv')
```

```python
gaiadr3.head()
```

```python
plt.figure(figsize=(10, 8))
plt.scatter(gaiadr3.ra, gaiadr3.dec, s=2.5, color='tomato', label='Gaia DR3$\pm2^\circ$')
plt.scatter(lit.ra, lit.dec, marker='D', color='black', label='Literature')
plt.gca().set_xlabel('$\\alpha$ [deg]', fontsize=16)
plt.gca().set_ylabel('$\delta$ [deg]', fontsize=16)
plt.legend();
```
